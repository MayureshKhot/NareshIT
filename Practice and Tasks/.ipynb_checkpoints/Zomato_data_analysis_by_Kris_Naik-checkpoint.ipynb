```python
# Clean 'Has Table booking' column
df['Has Table booking'] = df['Has Table booking'].apply(lambda x: str(x).strip().lower())

# Group by table booking option and calculate average rating
table_booking_rating = df.groupby('Has Table booking')['Aggregate rating'].mean().reset_index()

print("Average rating for restaurants with/without table booking:")
print(table_booking_rating)

# Visualize
plt.figure(figsize=(6,4))
sns.barplot(data=table_booking_rating, x='Has Table booking', y='Aggregate rating', palette='viridis')
plt.title('Average Rating vs Table Booking Availability')
plt.ylabel('Average Rating')
plt.xlabel('Has Table Booking')
plt.show()

# -----------------------------
# Insight 2: Online delivery vs price range
# -----------------------------
# Clean 'Has Online delivery' column
df['Has Online delivery'] = df['Has Online delivery'].apply(lambda x: str(x).strip().lower())

# Group by online delivery option and calculate average price range
online_delivery_price = df.groupby('Has Online delivery')['Price range'].mean().reset_index()

print("Average price range for restaurants with/without online delivery:")
print(online_delivery_price)

# Visualize
plt.figure(figsize=(6,4))
sns.barplot(data=online_delivery_price, x='Has Online delivery', y='Price range', palette='magma')
plt.title('Price Range vs Online Delivery Availability')
plt.ylabel('Average Price Range')
plt.xlabel('Has Online Delivery')
plt.show()
```

```python
# Cuisines Vs Ratings
```

```python
# Clean data: remove rows with missing cuisines or ratings
final_df = final_df.dropna(subset=['Cuisines', 'Aggregate rating'])

# For multi cuisine entries, only take first cuisine listed
final_df['Primary Cuisine'] = final_df['Cuisines'].apply(lambda x: x.split(',')[0].strip())

# Group by primary cisines and calculate average rating
cuisine_rating = final_df.groupby('Primary Cuisine')['Aggregate rating'].mean().reset_index()

# Sort to find lowest and highest rated cuisines
lowest_rated = cuisine_rating.sort_values(by='Aggregate rating').head(5)
highest_rated= cuisine_rating.sort_values(by='Aggregate rating', ascending=False).head(5)

print("Lowest rated cuisines:")
print(lowest_rated)

print("Highest rated cuisines:")
print(highest_rated)

# Visualize

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.barplot(data=highest_rated, y='Primary Cuisine', x='Aggregate rating', palette='Greens_r')
plt.title('Top 5 Highest rated cuisines')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')

plt.subplot(1,2,2)
sns.barplot(data=lowest_rated, y='Primary Cuisine', x='Aggregate rating', palette='Reds')
plt.title('Top 5 Lowest Rated Cuisines')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')

plt.tight_layout()
plt.show()
```

```python
# Check country codes to confirm India's code
print(df['Country Code'].unique())

# Filter data for India
india_df = df[df['Country Code'] == 1]  # Replace 1 with the actual country code for India if different

# Clean data: remove rows with missing cuisines or ratings
india_df = india_df.dropna(subset=['Cuisines', 'Aggregate rating'])

# For multi-cuisine entries, only take the first cuisine listed
india_df['Primary Cuisine'] = india_df['Cuisines'].apply(lambda x: x.split(',')[0].strip())

# Group by primary cuisine and calculate average rating
cuisine_rating_india = india_df.groupby('Primary Cuisine')['Aggregate rating'].mean().reset_index()

# Sort to find lowest and highest rated cuisines in India
lowest_rated_india = cuisine_rating_india.sort_values(by='Aggregate rating').head(10)
highest_rated_india = cuisine_rating_india.sort_values(by='Aggregate rating', ascending=False).head(10)

print("Top 10 Highest Rated Cuisines in India:")
print(highest_rated_india)

print("\nTop 10 Lowest Rated Cuisines in India:")
print(lowest_rated_india)

# Visualise
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.barplot(data=highest_rated_india, y='Primary Cuisine', x='Aggregate rating', palette='Greens_r')
plt.title('Top 10 Highest Rated Cuisines in India')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')

plt.subplot(1,2,2)
sns.barplot(data=lowest_rated_india, y='Primary Cuisine', x='Aggregate rating', palette='Reds')
plt.title('Top 10 Lowest Rated Cuisines in India')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')

plt.tight_layout()
plt.show()
```
