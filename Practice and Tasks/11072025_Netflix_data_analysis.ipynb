```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\NareshIT_PrakashSenapati\My Projects\Data\netflix_titles.csv")
df.head()
df.info()
```

```python
# check for missing values for all columns
df.isnull().sum()
```

```python
# Find the unique categories in type, rating and country

print(len(df['type'].unique()))
print(len(df['rating'].unique()))
print(len(df['country'].unique()))
```

```python
# Filling the null data with value 'Unknown'

df['rating'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['director'].fillna('Unknown', inplace=True)

```

```python
# conver the data_added into datetime format

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month for trend analysis later
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df.head()
```

```python
# Standardizing string columns:
# Remove the extra space and convert to lowers string for consistency.

df['type'] = df['type'].str.strip().str.lower()
df['rating'] = df['rating'].str.strip().str.lower()
```

```python
# Movies Vs TV shows count

type_counts = df['type'].value_counts()
print(type_counts)
```

```python
top_countries = df['country'].value_counts().head(10)
print(top_countries)
```

```python
# 'listed_in' contains genres, but multiple per entry, so split and count them

from collections import Counter

#drop missing values in listed in column

genres_series = df['listed_in'].dropna()

# Created empty list to store all genres
all_genres = []

# Loop through in row generes_series

for genre in genres_series:
    # Split the string by , and space
    split_genre=genre.split(', ')

    # Add each genre to the all_genres list
    for i in split_genre:
        all_genres.append(i)

genre_counts = Counter(all_genres)

# Conver to dataframe to get top 10 genres
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index', columns=['count'])
genre_df= genre_df.sort_values('count', ascending=False)

print(genre_df.head(10))
```

```python
# Group by Year added

content_trend = df.groupby('year_added').size()
print(content_trend)
```

```python
# Average duration of moves

#filer only movies
movies_df = df[df['type']=='movie']

# Duration column has min string, remove it and convert it to int
movies_df['duration_mins'] = movies_df['duration'].str.replace(' min', '', regex=False)
movies_df['duration_mins'] = pd.to_numeric(movies_df['duration_mins'], errors = 'coerce')

# Find the average duration
avg_duration = movies_df['duration_mins'].mean()
print(avg_duration)
```

```python
# Count directors with most titles

top_directors = df['director'].dropna().value_counts().head(10)
print(top_directors)
```

```python
# Visualizations

# Movies vs TV Shows
sns.countplot(data=df, x='type')
plt.title('Number of Movies Vs TV shows')
plt.show()
```

```python
# Find the top 10 Genres

sns.barplot(x=genre_df.head(10).index, y=genre_df.head(10)['count'])
plt.xticks(rotation=45)
plt.title('Top 10 Genres on Netflix')
plt.show()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing values hitmap')
plt.show()
```

```python
# Line plot, content added over the years

sns.lineplot(x=content_trend.index, y=content_trend.values)
plt.title('Content added per year')
plt.xlabel('Year')
plt.ylabel('Number of title added')
plt.show()
```

```python
# Ratings distribution pie chart

ratings = df['rating'].value_counts().head(10)
plt.pie(ratings, labels=ratings.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 ratings distributions')
plt.axis('equal')
plt.show()
```

```python
# Analyzing Netflix's rating column in depth

# Group rating manually

def categorize_rating(rating):
    kids=['tv-y', 'tv-y7', 'tv-g', 'g']
    teens=['pg', 'tv-pg', 'pg-13', 'tv-14']
    adults=['r', 'tv-ma', 'nc-17', 'nr']

    if rating in kids:
        return 'Kids'

    elif rating in teens:
        return 'Teens'

    elif rating in adults:
        return 'Adults'

    else:
        return 'Others'

# Apply categorization
df['audience_category'] = df['rating'].apply(categorize_rating)

# Count each category
audience_counts = df['audience_category'].value_counts()

print(audience_counts)

#Plot

plt.pie(audience_counts, labels=audience_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Netflix content distribution by Audience Category')
plt.axis('equal')
plt.show()
```
