```python
df_test = pd.read_csv(r"D:\NareshIT_PrakashSenapati\My Projects\Data\Black Friday sales\test.csv")
df_test.head()
```

```python
# Merge both train and test data
# What is the diffrence between these methods?

# pd.merge() -> Merges 2 dfs based on the common keys, like sql joints
# pd.concat() -> has different syntax, attaches 2 dfs on top of each other (rows) or side by side (columns)
# pd.append() -> It's deprecated
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
## Basic code
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis=1, inplace=True) # axis = 1 meands drop column wise
```

```python
df.head()
```

```python
# Data Preprocessing
pd.get_dummies(df['Gender'])
```

```python
# map function converts the values using dictionary
df['Gender']=df['Gender'].map({'F':0, 'M':1})
df.head()
```

```python
# Handle categorical feature - Age
df['Age'].unique()
# Here also we'll need to do encoding.

# wE can also do pd.get_dummies(df['Age'], drop_first=True)
# But here, we need to give ranks to these, like 0,1,2,3,4
# One hot encoder is not a very good practice at all, cos a lot of memroy consumption
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
# fixing categorical City_Category
df_city=pd.get_dummies(df['City_Category'], drop_first=True)
df_city.head()
```

```python
pd.concat([df,df_city],axis=1)
df.head()
```

```python
#
df.drop('City_Category', axis=1, inplace=True)
df.head()
```

```python
# missing values
df.isnull().sum()
# with purchase there is not problem, because that is test data mostly
```

```python
# this will become discrete feature (means what?)
df['Product_Category_2'].unique()
# In descreate feature, best way to treat null value is mode.
# No mean, cos it'll create a new categor, and we don't want that.
```

```python
df['Product_Category_2'].value_counts()
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
# Product category 3 replacing missing values.
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```
