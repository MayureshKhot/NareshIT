**Use Case/Features**

### Libraries and uses
1. Numpy- Nd Array
2. Matplotlib- Visualization
3. Pandas- Data Frame
4. Seaborn - Statistical visualization
5. Scipy- Statistical Calculations

## Methods in Pandas:
1. .describe(): gives description of the numerical data only. such as: std (sigma), mean, median, mode, percentile, quartile, etc.
2. .drop(): to delete the column from the existing dataset.
3. .head(): gives first 5 values
4. .tail(): gives last 5 values
5. transpose(): converts rows to columns (In pandas, the transpose() method is used to switch rows and columns of a DataFrame or Series.)

# Outliers
- In statistics outlier is the datapoint which is very far from other observations
- Outlier also called as ANOMALY DETECTION
- Outliers affect the analysis seriously.
- It's handled by logistic regression and naive baised algo

```python
vis5=sns.lmplot(data=df, x='InternetUsers', y='BirthRate')
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df, x='InternetUsers', y='BirthRate', fit_reg=False) # disappears the regression line, it's true by default.
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df, x='InternetUsers', y='BirthRate', fit_reg=True, hue='IncomeGroup') 
plt.show(vis5)
```
