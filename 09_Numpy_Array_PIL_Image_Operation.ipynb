# Numpy

* **Numpy** is a Python library used for numerical and scientific computing.
* It mainly works with **arrays**, specifically **ndarrays (n-dimensional arrays)**, which are multi-dimensional and homogeneous (same datatype).
* **Arrays** are collections of similar datatypes, stored in contiguous memory, making operations fast and efficient.

### Why do we use Numpy?

* All unstructured data such as **images, videos, and audio** are ultimately represented as arrays (pixel values, waveform values, etc.).
* Numpy provides powerful tools to handle and process this data efficiently.

### Using Numpy, we can perform:

* Arithmetic operations (addition, subtraction, multiplication, division)
* Statistical operations (mean, median, std, var, etc.)
* Bitwise operations
* Copying and viewing arrays efficiently
* Stacking (vertical and horizontal)
* Matrix operations
* Linear algebra (dot product, inverse, determinant, eigenvalues, etc.)
* Broadcasting (performing operations on arrays of different shapes)
* Mathematical operations (trigonometric, logarithmic, exponential, etc.)
* Searching, sorting, and filtering

ðŸ”‘ **Note:** Indexing in Numpy arrays also starts from **0**.

## Difference between List and Numpy Array

| **List**                                                              | **Numpy Array**                                        |
| --------------------------------------------------------------------- | ------------------------------------------------------ |
| Can store **heterogeneous** datatypes (e.g. int, str, float together) | Stores **homogeneous** datatype elements only          |
| Cannot perform **vectorized operations** directly (uses loops)        | Supports **vectorized operations**, making it faster   |
| Takes **more memory** for large datasets                              | Takes **less memory** due to optimized fixed datatypes |
| General purpose data structure                                        | Specifically used for **scientific computation**       |
| No built-in multi-dimensional support                                 | Built-in **multi-dimensional array** support           |

ðŸ”Ž **Correction:** Arrays take **less memory** compared to lists because of their fixed datatype and contiguous allocation.

## Important Numpy Functions

1. **np.arange()**

   * Creates an array with evenly spaced values within a given interval.
   * Syntax: `np.arange(start, stop, step)`
   * At least one argument (stop) is required.
   * Example: `np.arange(1,10,2)` âž” \[1 3 5 7 9]
   * If start > stop and step is positive âž” returns empty array.

2. **np.zeros()**

   * Creates an array filled with zeros.
   * Syntax: `np.zeros(shape, dtype=float)`
   * Example: `np.zeros((3,3), dtype=int)`
   * Useful for initializing arrays before filling with data (e.g. black images).

3. **np.ones()**

   * Similar to zeros(), but creates an array filled with ones.
   * Syntax: `np.ones(shape, dtype=float)`
   * Example: `np.ones((2,4), dtype=int)`

4. **np.eye()**

   * Creates an identity matrix.
   * Syntax: `np.eye(N)` where N is number of rows and columns.

5. **np.linspace()**

   * Returns evenly spaced numbers over a specified interval.
   * Syntax: `np.linspace(start, stop, num=50)`

6. **np.random.rand() / np.random.randn() / np.random.randint()**

   * `rand()` âž” random floats between 0 and 1.
   * `randn()` âž” random samples from standard normal distribution.
   * `randint()` âž” random integers within a range.

7. **np.reshape()**

   * Changes the shape of an existing array without changing data.

## Slicing in Matrix

- arr[2:10] --> print the row from 2nd row to n-1 i.e 9th row
- arr[2,10] --> print 10th element in second row.

```python
b=np.random.randint(10,20,(5,4))
b
```

```python
b[:]
```

```python
b[1:4] # Gives first 3 rows as output. 1 is inclusive, 4 is exclusive
```

```python
b[-1:] # -1 gives the last row only
```

```python
b[:-1] #Why did it skip the last row?
# Because of n-1 formula, -1-1 = -2. Therefrore it's giving output from last second i.e -2 row.
```

```python
b
```

```python
b[1,2] # returns 2nd row, 3rd element
# So comma is used to get specific numbers, wehreas : slicing is used to return entire rows.
```
