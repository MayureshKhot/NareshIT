# 1. Dot notation (.)
## ‚úÖ When it works:

- When the column name is a valid Python attribute name, i.e.,

- No spaces

- No special characters

- Doesn‚Äôt clash with DataFrame methods (e.g. count, sum).

## üî¥ When it fails:

- If column name has spaces, like Movie Title

- If column name conflicts with pandas methods, like count or sum

# 2. Bracket notation ([])
- ‚úÖ When to use:

- Always works, regardless of:

- Spaces in column names

- Special characters

- Conflicts with method names

**‚úîÔ∏è Prefer bracket notation ([]) for production and general use to avoid hidden errors.**
