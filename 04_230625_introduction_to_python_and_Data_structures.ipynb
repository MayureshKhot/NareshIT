✅ Corrected Definitions:
Interpreter:
An interpreter reads and executes code line by line. It translates each high-level instruction into machine code and runs it immediately, without converting the whole program at once.
🧠 Example: Python is typically interpreted using the CPython interpreter.

Compiler:
A compiler translates the entire source code into machine code (or bytecode) in one go before any part of the program is executed. The result is a separate executable file or intermediate code.
🧠 Example: C and C++ use compilers like GCC.

⚠️ Common Misunderstanding:
Saying "executing at once" vs "line by line" oversimplifies the process.

Python is often said to be "interpreted," but tools like PyPy or the CPython bytecode compiler involve both compilation and interpretation phases.

🔄 In Python's Case:
Python compiles source code (.py) into bytecode (.pyc) using a built-in compiler.

That bytecode is then interpreted by the Python virtual machine (PVM).
